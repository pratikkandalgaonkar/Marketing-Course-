```python
train_df = pd.get_dummies(traindf, columns=['V1','V4','V5','V44','V6'], drop_first = True)
train_df
```

```python
test_df = pd.get_dummies(testdf, columns = ['V1','V4','V5','V44','V6'], drop_first = True)
test_df
```

```python
x_train = train_df.drop(['V86','V6_8'], axis = 1)
y_train = train_df['V86']
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(fit_intercept = True)
```

```python
param = {'penalty':['l1','l2'],
        'C':np.linspace(0.01,0.06,100),
        'class_weight':['balanced',None]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_cv = GridSearchCV(model,cv = 10,param_grid = param, n_jobs = -1, verbose = 20,scoring = 'roc_auc')
```

```python
grid_cv.fit(x_train,y_train)
```

```python
grid_cv.best_estimator_
```

```python
train_score = grid_cv.best_estimator_.predict_proba(x_train)[:,1]
```

```python
cutoffs = np.linspace(0.001,1,800)
```

```python
from sklearn.metrics import fbeta_score
```

```python
fbeta = []
```

```python
for cutoff in cutoffs:
    predicted = (train_score>cutoff).astype(int)
    fbeta.append(fbeta_score(y_train,predicted,2))
```

```python
my_cutoff=cutoffs[fbeta==max(fbeta)]
```

```python
my_cutoff
```

```python
predictions=(grid_cv.predict_proba(test_df)[:,1]>my_cutoff).astype(int)
```

```python
pd.Series(predictions).value_counts()
```

```python
submissions=pd.DataFrame({'V86':predictions})
```
